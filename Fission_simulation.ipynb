```python
import random
import numpy as np
from math import cos, pi, sqrt, acos, sin
from numpy import random
```

```python
####### 1D MODEL ######


L = 0.1           
a = 0.017         
b = 0.21           
R = np.sqrt(2 * a * b)  

N_initial = 100    
neutrons_per_fission = 2  

secondary_count = 0

for i in range(N_initial):
    
    x0 = L * random.random()
    
    for j in range(neutrons_per_fission):
        
        if random.random() > 0.5:
            x_new = x0 + R
        else:
            x_new = x0 - R
        
        if 0 <= x_new <= L:
            secondary_count += 1

print("Total secondary fissions:", secondary_count)
print("Average secondary fissions per initial fission:", secondary_count / N_initial)


def neutrons():
    """Number of secondary neutrons produced in each fission.

    Returns an integer number of neutrons, with average 2.5."""
    i=int(random.normal()+3.0)
    if (i<0): return 0
    else: return i

def diffusion():
    """Distance diffused by a neutron before causing fission.

    Returns a random number with probability density p(s) =
    s^2 exp(-3s^2/R^2). This distribution has a mean of 1, so
    multiply by R to get the physical distance."""
    a=cos(2.0*pi*random.random())
    return sqrt(-0.667*(log(random.random())+log(random.random())*a*a))


lengths = [0.1, 0.12, 0.14, 0.16, 0.18, 0.2]  

N_variable = 5000
 
for L in lengths:
    secondary_total = 0
    for i in range(N_variable):
        x0 = L * random.random()
        n_neutrons = neutrons()
        for j in range(n_neutrons):
            x_new = x0 + R if random.random() > 0.5 else x0 - R
            if 0 <= x_new <= L:
                secondary_total += 1
    avg_per_fission = secondary_total / N_variable
    print(f"L = {L:.3f} m, Average secondary fissions per initial fission: {avg_per_fission:.3f}")


Ls = np.linspace(0.14, 0.16, 201)
avg = []

for L in Ls:
    secondary_total = 0
    for i in range(N_variable):
        x0 = L * random.random()
        n = neutrons()
        for j in range(n):
            x_new = x0 + R if random.random() > 0.5 else x0 - R
            if 0 <= x_new <= L:
                secondary_total += 1
    avg.append(secondary_total / N_variable)

# Find which length gives average closest to 1.0
min_diff = 1000         
best_index = 0         
for i in range(len(avg)):
    diff = abs(avg[i] - 1)  
    if diff < min_diff:    
        min_diff = diff     
        best_index = i      

print("\nCritical L =", Ls[best_index])  

```

```python
####### 3D MODEL ######


density = 18.7  # Mg/m³
L = 0.1
N_initial = 1000
secondary_count_3d = 0

for i in range(N_initial):

    x0 = L * random.random()
    y0 = L * random.random() 
    z0 = L * random.random()
    
    
    for j in range(neutrons()):
        s = diffusion() * R
        
        phi = 2.0 * pi * random.random()      # longitude
        theta = acos(2.0 * random.random() - 1.0)  # colatitude
        
        x_new = x0 + s * sin(theta) * cos(phi)
        y_new = y0 + s * sin(theta) * sin(phi)
        z_new = z0 + s * cos(theta)
        
        if (0 <= x_new <= L and 0 <= y_new <= L and 0 <= z_new <= L):
            secondary_count_3d += 1

print(f"Using L = {L:.3f} m, and N = {N_initial} to test model")
print("Total secondary fissions in 3D:", secondary_count_3d)
print("Average secondary fissions per initial fission:", secondary_count_3d / N_initial)
print()


lengths_3d = [0.10, 0.12, 0.14, 0.16, 0.18, 0.20]  
N_variable_3d = 5000

for L in lengths_3d:
    secondary_total_3d = 0
    for i in range(N_variable_3d):
        x0 = L * random.random()
        y0 = L * random.random()
        z0 = L * random.random()
        
        n = neutrons()
        for j in range(n):
            s = diffusion() * R
        
            phi = 2.0 * pi * random.random()
            theta = acos(2.0 * random.random() - 1.0)
            
            x_new = x0 + s * sin(theta) * cos(phi)
            y_new = y0 + s * sin(theta) * sin(phi)
            z_new = z0 + s * cos(theta)
            
            if (0 <= x_new <= L and 0 <= y_new <= L and 0 <= z_new <= L):
                secondary_total_3d += 1
                
    avg_per_fission_3d = secondary_total_3d / N_variable_3d
    volume = L**3
    mass = volume * density * 1000
    print(f"L = {L:.3f} m, Average secondary fissions: {avg_per_fission_3d:.3f}, Mass: {mass:.3f} kg")


Ls_3d = np.linspace(0.14, 0.16, 201)
avg_3d = []

for L in Ls_3d:
    secondary_total = 0
    for i in range(N_variable_3d):
        x0 = L * random.random()
        y0 = L * random.random()
        z0 = L * random.random()
        
        n = neutrons()
        for j in range(n):
            s = diffusion() * R
            
            phi = 2.0 * pi * random.random()
            theta = acos(2.0 * random.random() - 1.0)
            
            x_new = x0 + s * sin(theta) * cos(phi)
            y_new = y0 + s * sin(theta) * sin(phi)
            z_new = z0 + s * cos(theta)
            
            if (0 <= x_new <= L and 0 <= y_new <= L and 0 <= z_new <= L):
                secondary_total += 1
                
    avg_3d.append(secondary_total / N_variable_3d)

min_diff_3d = 1000
best_index_3d = 0
for i in range(len(avg_3d)):
    diff = abs(avg_3d[i] - 1)
    if diff < min_diff_3d:
        min_diff_3d = diff
        best_index_3d = i

critical_L = Ls_3d[best_index_3d]
critical_volume = critical_L**3
critical_mass = critical_volume * density * 1000

print(f"\nCritical L for 3D model = {critical_L:.3f} m")
print(f"Critical mass: {critical_mass:.3f} kg")
```
